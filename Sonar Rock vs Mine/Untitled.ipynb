```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Data Processing

```python
sonar_data = pd.read_csv("sonar data.csv")
```

```python
sonar_data.groupby(['R']).mean()
```

```python
# separating data and labels
X = sonar_data.drop(columns='R', axis=1)
Y = sonar_data['R']
```

```python
print(X)
```

```python
print(Y)
```

```python
print('X-shape : ',X.shape," Y-shape : ",Y.shape)
```

```python
model = LogisticRegression()
```

```python
X_train, X_, Y_train, Y_ = train_test_split(X,Y, test_size = 0.40, random_state = 5)
```

```python
print(f"X_train : {X_train.shape} X_ : {X_.shape} Y_train : {Y_train.shape} Y_ {Y_.shape}")
```

```python
X_cv, X_test, Y_cv, Y_test = train_test_split(X_,Y_, test_size = 0.50, random_state = 8)
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy Score : ', training_data_accuracy)
```

```python
#Accuracy on cross validation data
X_cv_prediction = model.predict(X_cv)
cv_data_accuracy = accuracy_score(X_cv_prediction, Y_cv)
```

```python
print('Accuracy Score : ', cv_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy Score : ', test_data_accuracy)
```
